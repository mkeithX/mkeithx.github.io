### Widgets

```python
@interact(x=(0, 32))

def square(x):
    print("The square of {} is {}".format(x, x**2))
```

```python
@interact(frequency=(1, 10, 1), amplitude=(1, 5, 0.1))
def plot_sine_wave(frequency, amplitude):
    x = np.linspace(0, 10, 1000)
    y = amplitude * np.sin(frequency * x)
    plt.plot(x, y)
    plt.title('Sine Wave')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.show()
```
