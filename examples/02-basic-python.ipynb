### ðŸ’¡For statements

```python
## List:
fruits = ['apple', 'banana', 'orange']

for fruit in fruits:
    print(fruit)
```

```python
# Range:
for i in range(5):
    print(i)
```

```python
# String:
word = "Jupyter Notebook"

for char in word:
    print(char, end=" ")
```

```python
# over Dictionary Keys
person = {'name:': 'Johanna', 'Age:': 40, 'City:': 'Cebu'}
for key in person:
    print(key, person[key])
```

```python
# Enumerate for Index and Value:
colors = ['red', 'green', 'blue']
for index, color in enumerate(colors):
    print(f"Index: {index}, Color: {color}")
```

### ðŸ’¡If statements

```python
# Conditional Check:
x = 10
if x > 5:
    print("x is greater than 5")
```

```python
# If-Else Statement:

temperature = 25
if temperature > 30:
    print("It's hot outside.")
else:
    print("It's not too hot.")
```

```python
# Nested If Statements:
num = 15
if num > 10:
    print("Number is greater than 10")
    if num % 2 == 0:
        print("Number is even")
    else:
        print("Number is odd")
```

```python
# Using Elif for Multiple Conditions:
grade = 75
if grade >= 90:
    print("A")
elif 80 <= grade < 90:
    print("B")
elif 70 <= grade < 80:
    print("C")
else:
    print("Below C")
```

```python
# Checking Membership with `in``:
fruits = ['apple', 'banana', 'orange']
if 'banana' in fruits:
    print("Banana is in the list")
else:
    print("Banana is not in the list")
```

### ðŸ’¡While statements

```python
# Basic `while` Loop:

count = 0
while count < 5:
    print(count, end=" ")
    count += 1
```
